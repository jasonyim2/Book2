```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
```

```python
# Document 1 감성 분석
classifier("4,000 a day dying from the so called Covid-19 vaccine.")
```

```python
# Document 8과 10 감성 분석
results = classifier(["Thanks to Dr. @Swamy39 jee & Health Minister, Dr. Harsh Vardhan jee for promoting \
                       India's first indigenous Covid Vaccine from Bharat Biotech.", 
                      "The day after the #Moderna #COVID19Vaccine... My arm is a little sore and \
                       I am a little more fatigued than usual. \
                       My grandparents had no noticeable effects though."])
for result in results:
    print(f"label: {result['label']}, with score: {round(result['score'], 4)}")
```

```python
# 트랜스포머 모델명 지정
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"

# AutoTokenizer 및 AutoModelForSequenceClassification을 import
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# 지정한 트랜스포머 모델에서 사전훈련시킨 AutoModelForSequenceClassification을 불러와 model이라는 변수에 저장
# 지정한 트랜스포머 모델에서 사전훈련시킨 AutoTokenizer를 불러와 tokenizer라는 변수에 저장
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
# Document 8을 스페인어로 번역후 투입
classifier = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
classifier("Gracias al Dr. @Swamy39 jee y al Ministro de Salud, \
            Dr. Harsh Vardhan jee por promover la primera vacuna autóctona \
            contra el covid de la India de Bharat Biotech.")
```
